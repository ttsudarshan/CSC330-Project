```python
circuit = circuit.drop(columns=['name','previous_names','direction','latitude','longitude'])
circuit = circuit.rename(columns = {
    'total_races_held' : 'race_count'
})
```

## Analysis

```python
tracks = ['monza', 'monaco', 'silverstone'] # Circuits of interest; have the most amount of racing data to analyze
```

```python
monza_races = race.query(f'circuit_id == "{tracks[0]}"')

monza_races = monza_races.drop(columns=['time','grand_prix_id', 'circuit_id', 'circuit_type', 'direction']) # Remove redundant information (same for all rows)
monza_races = monza_races.dropna(axis=1, how='all') # Remove non-applicable information (null for all rows)
monza_races
```
